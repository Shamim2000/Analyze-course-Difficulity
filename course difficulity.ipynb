#Since only the ETP_100 columns is Vacant We will deal with that columns by inserting number which are desrving.


```python
def preprocessing(cols):
    a,b,c,d = cols[0],cols[1],cols[2],cols[3]
    if(pd.isnull(d)):
        return (a+b+c)//3
    else:
        return d
```

```python
dataset['ETP_100'] = dataset[['CA_100', 'MTT_50', 'ETT_100', 'ETP_100']].apply(preprocessing,axis = 1)
```

```python
dataset['total_marks'] = marks
```

```python
def grades (cols):
    ca,mte,etp_theory,etp_practical,courseattendace = cols[0],cols[1],cols[2],cols[3],cols[4]
    ca = (ca/100)*25
    courseattendace = (courseattendace/100)*25
    total = ((ca+mte+etp_theory+etp_practical+etp_theory+courseattendace)/200)*100
    return total
```

```python
dataset['total_marks'] = dataset[['CA_100','MTT_50','ETT_100','ETP_100','Course_Att']].apply(grades,axis = 1)
```

```python
grade =[0]*dataset.shape[0]
```

```python
dataset["grades"] = grade
```

```python
m = max(dataset['total_marks'])
grade = []
```

```python
def marks (cols):
    mark = (cols[0]/m)*100
    
    if((mark>95)):
        return 1;
    elif(mark>90):
        return 2
    elif(mark>83):
        return 3
    elif(mark>75):
        return 4
    elif(mark>60):
        return 5
    elif(mark>54):
        return 8
    elif(mark>33):
        return 9
    else:
        return 10
    
```

```python
dataset['NUMgrade'] = dataset[['total_marks']].apply(marks,axis = 1)
```

```python
dataset.drop("Grade",axis = 1,inplace= True)
```

```python
dataset.reset_index(drop=True,inplace=True)
dataset.head()
```

```python
dummyDF = pd.get_dummies(dataset.Course)
```

```python
remainingDF = dataset.iloc[:,1:]
```

```python
data2 = pd.concat([ dummyDF , remainingDF ] , axis = 1)
data2
```

```python
x = data2.iloc[:,:-2]
x
```

```python
y = data2.iloc[:,-1]
y
```

```python
#Use algorithms to get the view.
from sklearn.cluster import KMeans
model = KMeans(n_clusters=5,
              random_state=10)
```

```python
model.fit(x)
```

```python
Course_difficulity = model.predict(x)
```
